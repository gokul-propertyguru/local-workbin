```python
import psycopg2
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
sql = ('SELECT id_dw_listing,'
       'EXTRACT(year FROM DATE) "year",'
       'EXTRACT(week FROM DATE) "week",'
       'SUM(unique_views) "Unique Views",'
       'SUM(views) "Views"'
       'FROM dw_reports.fact_views '
       'where EXTRACT(year FROM DATE) = 2015 and EXTRACT(month FROM DATE)>=1 and substring(id_dw_listing from 1 for 2)=\'SG\' '
       'GROUP BY id_dw_listing,year,week '
       'ORDER BY year DESC,week DESC'
       )
col_list_sql = "select column_name, data_type, character_maximum_length from INFORMATION_SCHEMA.COLUMNS where table_name = 'fact_views'"
```

```python
def create_conn():
    conn = psycopg2.connect(
        host="pg-redshift-cluster.cgnifbeifavu.ap-southeast-1.redshift.amazonaws.com", 
        user="rsreadonly", 
        port=5439, 
        password="Pguru123", 
        dbname="dev")
    return conn
    
def execute_sql(connection,stmt):
    cur = connection.cursor()
    cur.execute(stmt)
    result = cur.fetchall()
    cur.close()
    return result

def createDF(col_list,data):
    #data [(col_list),(col_list),...(col_list)]
    df = pd.DataFrame(data=data,columns=col_list)
    return(df)
```

```python
conn = create_conn()
result = execute_sql(conn,sql)
```
