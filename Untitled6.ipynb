```python
shp = fiona.open("subzone//Subzone_Census2010.shp")
bds = shp.bounds
shp.close()
cv = SVY21.SVY21()

extra = 0.01
ll = cv.computeLatLon(bds[1],bds[0])
ur = cv.computeLatLon(bds[3],bds[2])
coords = list(chain(ll,ur))
w,h = coords[2]-coords[0],coords[3]-coords[1]
w,h
```

```python
print ll
print ur
```

```python
m = Basemap(
    projection = 'tmerc',
    lon_0=103.49,
    lat_0=1.21,
    ellps='WGS84',
    llcrnrlat=coords[0] - extra * w,
    llcrnrlon=coords[1] - extra + 0.01 * h,
    urcrnrlat=coords[2] + extra * w,
    urcrnrlon=coords[3] + extra + 0.01 * h,
    lat_ts=0,
    resolution='h',
    suppress_ticks=True   
)

m.readshapefile(
    "subzone//Subzone_Census2010WGS84",
    "subzone",
    color=None,
    zorder=2
)
```

```python
from shapely.geometry import Point,Polygon,MultiPoint,MultiPolygon
from shapely.prepared import prep
import pandas as pd
df_map = pd.DataFrame({
        'subzone_name':[ward['SUBZONE_N'] for ward in m.subzone_info],
        'poly':[Polygon(xy) for xy in m.subzone]})
df_map['area_m']=df_map['poly'].map(lambda x: x.area)
df_map['area_km']=df_map['area_m']/100000
```

```python
df_map.to_csv("subzone_map.csv",index=False)
```
